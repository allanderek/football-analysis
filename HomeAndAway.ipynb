```python
def scatter_home_away_points_column(data_frame, team, column):
    home_team_data_frame = data_frame[data_frame.HomeTeam == team]
    home_points = home_team_data_frame[['HomePoints']].sum()
    away_team_data_frame = data_frame[data_frame.AwayTeam == team]
    away_points = away_team_data_frame[['AwayPoints']].sum()
    total_points = home_points + away_points
    
    home_mean_row = home_team_data_frame[['Home' + column]].mean()
    away_mean_row = away_team_data_frame[['Away' + column]].mean()
    
    column_mean = home_mean_row + away_mean_row
    
    plot.title(column)
    plot.xlabel('Home Points')
    plot.ylabel('Away Points')
    plot.scatter(total_points, column_mean)
    plot.annotate(team,
        xy = (total_points, column_mean), xytext = (-20, 20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))

for column in columns_of_interest:
    for team in epl_league.teams:
        scatter_home_away_points_column(epl_league.data_frame, team, column)
    plot.show()
    plot.close()
```
