```python
import pandas
from IPython.display import display, HTML
import urllib.request
```

```python
class League(object):
    def __init__(self, league_short_title, fixtures_directory):
        self.title = fixtures_directory # Probably should have keyword argument for this with this as default.
        data_file_basename = league_short_title + ".csv"
        self.data_url = "http://www.football-data.co.uk/mmz4281/1415/" + data_file_basename
        self.data_file = "data/1415/" + data_file_basename
        self.fixtures_url = "http://www.bbc.co.uk/sport/football/" + fixtures_directory + "/fixtures"
        self.fixtures_file = "data/1415/" + league_short_title + "-fixtures.html"
    
    def retrieve_data_and_fixtures(self):
        urllib.request.urlretrieve(self.data_url, self.data_file)
        urllib.request.urlretrieve(self.fixtures_url, self.fixtures_file)
    
    def display_title(self):
        display(HTML("<h2>" + self.title + "</h2>"))
```

```python
epl_league = League("E0", "premier-league")
leagues = [epl_league,
           League("E1", "championship"),
           League("E2", "league-one"),
           League("E3", "league-two"),
           League("SC0", "scottish-premiership"),
           # League("SC1", "scottish-championship") No shots data
           ]
```

```python
def get_home_pdo(row):
    """ 1000*(goals for/shots on target for + (shots on target against – goals against)/shots on target against) """
    goals_for = row.FTHG
    shots_on_target_for = row.HST
    shooting_percentage = goals_for / shots_on_target_for if shots_on_target_for else 0.0
    shots_on_target_against = row.AST
    goals_against = row.FTAG
    saves = shots_on_target_against - goals_against
    # The save percentage is basically the same as the 1.0 - the away team's shot percentage
    # which is why this is 1.0 in the case that there were no away shots.
    save_percentage = saves / shots_on_target_against if shots_on_target_against else 1.0
    return (shooting_percentage + save_percentage) - 1
def get_away_pdo(row):
    return 0.0 - row.HomePDO
def add_pdo_columns(data_frame):
    data_frame['HomePDO'] = data_frame.apply(get_home_pdo, axis=1)
    data_frame['AwayPDO'] = data_frame.apply(get_away_pdo, axis=1)
def get_home_tsott(row):
    """Shots on target for/Total shots for + Shots off target against/Total shots against"""
    shots_on_target_for = row.HST
    shots_for = row.HS
    shots_for_target_ratio = shots_on_target_for / shots_for if shots_for else 0
    shots_against = row.AS
    shots_off_target_against = shots_against - row.AST
    # Similar to the PDO, this is basically 1.0 - the away team's shots_target_ratio which
    # is again why this is 1.0 if there were no away shots.
    shots_against_off_target_ratio = shots_off_target_against / shots_against if shots_against else 1.0
    return (shots_for_target_ratio + shots_against_off_target_ratio) - 1.0
def get_away_tsott(row):
    return 0.0 - row.HomeTSOTT
def add_tsott_columns(data_frame):
    data_frame['HomeTSOTT'] = data_frame.apply(get_home_tsott, axis=1)
    data_frame['AwayTSOTT'] = data_frame.apply(get_away_tsott, axis=1)
def get_home_tsr(row):
    """Shots on target for/Total shots for + Shots off target against/Total shots against.
       However to keep it normalised around zero, with a negative value meaning bad and a
       positive value meaning good. So we take 0.5 from the ratio and multiply by 2.
    """
    shots_for = row.HS
    shots_against = row.AS
    total_shots = shots_for + shots_against
    tsr = shots_for / total_shots if total_shots else 0.5
    return tsr
def get_away_tsr(row):
    return 1.0 - row.HomeTSR
def add_tsr_columns(data_frame):
    data_frame['HomeTSR'] = data_frame.apply(get_home_tsr, axis=1)
    data_frame['AwayTSR'] = data_frame.apply(get_away_tsr, axis=1)
def get_team_rating(pdo, tsott, tsr):
    """ Essentially tsr * tsott * pdo, but not weight equally, James Grayson gives it as:
        Rating = (0.5+(TSR-0.5)*0.732^0.5)*(1.0+(%TSOTt-1.0)*0.166^2)*(1000+(PDO-1000)*0.176^0.5)
        But we have normalised the three values to average at 0. Note that by doing this we really
        shouldn't multipy. Instead we will add, but we will add only 0.82 of tsr, 0.45 of tsott and
        0.4 of pdo because these appear to be the repeatable portions.
    """
    normalised_tsr = (tsr - 0.5) * 2.0
    rating = (0.82 * normalised_tsr) + (0.45 * tsott) + (0.4 * pdo)
    return rating

def get_home_team_rating(row):
    return get_team_rating(row.HomePDO, row.HomeTSOTT, row.HomeTSR)
def get_away_team_rating(row):
    return get_team_rating(row.AwayPDO, row.AwayTSOTT, row.AwayTSR)
def add_team_rating_columns(data_frame):
    data_frame['HomeTeamRating'] = data_frame.apply(get_home_team_rating, axis=1)
    data_frame['AwayTeamRating'] = data_frame.apply(get_away_team_rating, axis=1)
def add_team_rating_related_columns(data_frame):
    add_pdo_columns(data_frame)
    add_tsott_columns(data_frame)
    add_tsr_columns(data_frame)
    add_team_rating_columns(data_frame)
rating_fields = ['PDO', 'TSOTT', 'TSR', 'TeamRating']
home_rating_fields = ["Home" + f for f in rating_fields]
away_rating_fields = ["Away" + f for f in rating_fields]
goal_fields = ['FTHG', 'FTAG']
```

Getting the team averages is a small point of contention. Suppose we wish to calculate the total-shots ratio for a particular team. We can either calculate the ratio in each game and then take the average of all games in which they are included. Or we can calculate add up their shots for and shots against totals for all games in which they are involved and then take the ratio. These are not equivalent. Imagine a team is involved in two games, in the first, they take 99 of the 100 shots in the game. In the second they take 1 of 2. If we take the TSR of both games it will 0.99 and 0.5, if we then take that average it is just below 0.75. But if we add up the shots in the first place we get 100 shots for out of 102 shots, which is a ratio of close to 1. It is arguable which is the correct way but I've gone for the former since intuitively it feels to me as if the team should not get a TSR of near 100% when they have a TSR of 50% in one of those two games. I have not checked, but suspect that this does not really make much difference over the course of the season because games have numbers of shots that are more comparable. My method however may skew things because games with very low shot numbers may have outlier shot ratios such as 0 or 1. It's hard to say, for example, if you take the only shot of the game, should you be rewarded highly with a TSR of 1, or barely at all with the addition of simply 0 shots against and 1 for?

It's actually really interesting because I suspect that if anything we'd want to reward a team *more* for winning the shots totals in a game with lots of shots than a game in which there are very few. In particular that one shot in the game may have been from an outlandish position.

```python
def get_team_averages(data_frame, team):
    home_team_data_frame = data_frame[data_frame.HomeTeam == team]
    home_mean_row = home_team_data_frame[home_rating_fields].mean()
    away_team_data_frame = data_frame[data_frame.AwayTeam == team]
    away_mean_row = away_team_data_frame[away_rating_fields].mean()
    return home_mean_row, away_mean_row
def get_full_data_frame(data_file):
    data_frame = pandas.read_csv(data_file)
    add_team_rating_related_columns(data_frame)
    return data_frame
```

```python
def compare_match_teams(data_frame, team1, team2):
    home_team_data_frame = data_frame[data_frame.HomeTeam == team1]
    home_mean_row = home_team_data_frame[home_rating_fields + goal_fields].mean()
    away_team_data_frame = data_frame[data_frame.AwayTeam == team2]
    away_mean_row = away_team_data_frame[away_rating_fields + goal_fields].mean()
    return home_mean_row, away_mean_row
```

```python
def compare_matches(matches, data_frame):
    for home, away in matches:
        home_stats, away_stats = compare_match_teams(data_frame, home, away)
        for rating_field in rating_fields:
            home_rating_field = "Home" + rating_field
            home_stats[rating_field] = home_stats[home_rating_field]
            away_rating_field = "Away" + rating_field
            away_stats[rating_field] = away_stats[away_rating_field]

        home_stats['Goals For'] = home_stats['FTHG']
        home_stats['Goals Against'] = home_stats['FTAG']
        away_stats['Goals For'] = away_stats['FTAG']
        away_stats['Goals Against'] = away_stats['FTHG']

        goal_fields = ['Goals For', 'Goals Against']
        return_data_frame = pandas.DataFrame()
        return_data_frame[home] = home_stats[rating_fields + goal_fields]
        return_data_frame[away] = away_stats[rating_fields + goal_fields]
        display(return_data_frame)
```

```python
team_aliases = {'Dundee Utd' : 'Dundee United',
                'Inverness CT' : 'Inverness C',
                'Partick Thistle' : 'Partick',
                'Man Utd' : 'Man United',
                'Sheff Wed' : 'Sheffield Weds',
                'Nottm Forest' : "Nott'm Forest",
                'Sheff Utd' : 'Sheffield United',
                'MK Dons' : 'Milton Keynes Dons',
                'Fleetwood' : 'Fleetwood Town',
                'Peterborough' : 'Peterboro',
                'Crawley' : 'Crawley Town'
               }
def alias_team(team):
    return team_aliases.get(team, team)

def get_match_teams(match_details):
    home_team_span = match_details.find('span', class_='team-home teams')
    home_team = home_team_span.a.contents[0]
    away_team_span = match_details.find('span', class_='team-away teams')
    away_team = away_team_span.a.contents[0]
    return (alias_team(home_team), alias_team(away_team))
    
def compare_fixtures(data_frame, fixtures_page):
    soup = BeautifulSoup(open(fixtures_page, encoding='utf-8'))
    
    match_details_list = soup.find_all('td', class_='match-details')
    matches = [get_match_teams(md) for md in match_details_list]
    compare_matches(matches[0:12], data_frame)
```

```python
def display_recent_matches(data_frame, n=10):
    match_fields = (['HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'HS', 'AS', 'HST', 'AST', 'HR', 'AR'] + 
                    home_rating_fields + away_rating_fields)
    selected_data = data_frame[match_fields].tail(n=n)
    reversed_data_frame = selected_data.iloc[::-1]
    # Partly just to remind myself how to do this
    reversed_data_frame = reversed_data_frame.rename(columns={'HS': 'HomeShots', 'AS': 'AwayShots',
                                                              'FTHG': 'HomeGoals', 'FTAG': 'AwayGoals'})
    return reversed_data_frame
```
